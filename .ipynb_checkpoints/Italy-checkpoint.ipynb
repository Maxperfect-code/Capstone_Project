# Let's create a function to do the same exercise in all the Italian's province.

First step is to find out all the latitudes and longitudes of all the province in all the country.

We can find all the province in the next website; whit a for cicle we'll find all the value we need and we create a new Dataframe

```python
url_prov= 'https://www.studiomadera.it/vari/195-province'
data=requests.get(url_prov).text
soup=BeautifulSoup(data,"html5lib")
table=soup.find_all('table')
```

```python
Dataframe= pd.read_html(str(table[1]), flavor='bs4')
Province=Dataframe[0]
Province.rename(columns={0:'Sigla',1:'Provincia',2:'Regione'},inplace=True)
Province=Province.loc[1:,:]
local=pd.DataFrame(columns=['Provincia','lat','long'])
Province_italiane=Province['Provincia']
```

```python
for address in Province_italiane:
        geolocator = Nominatim(user_agent="foursquare_agent")
        location = geolocator.geocode(address)
        latitude = location.latitude
        longitude = location.longitude
        local=local.append({'Provincia': address,
                            'lat': latitude,
                            'long': longitude},ignore_index=True)
```

We called 'local' our new dataset. Let's see the first 2 lines of it.

```python
local.head(2)
```

Now we create the function to find out all the best 100 venues for all the province

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    all_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    all_venues.columns = ['Provincia', 
                  'Provincia Latitude', 
                  'Provincia Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(all_venues)
```

```python
Italy_venues = getNearbyVenues(names=local['Provincia'],
                                   latitudes=local['lat'],
                                   longitudes=local['long']
                                  )
```

# Anylize each Province

The first step is to transform our cateogories into dummy

```python
# one hot encoding
Italy_dummy = pd.get_dummies(Italy_venues[['Venue Category']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
Italy_dummy['Provincia'] = Italy_venues['Provincia'] 
```

```python
move_columns = [Italy_dummy.columns[-1]] + list(Italy_dummy.columns[:-1])
Italy_dummy = Italy_dummy[move_columns]
Italy_dummy.head(2)
```

```python
Italy_dummy.shape
```

We now create a group by Province taking the mean of each category

```python
Italy_grouped=Italy_dummy.groupby('Provincia').mean().reset_index()
```

```python
Italy_grouped
```

```python
Italy_grouped.shape
```

Let's create a new Dataframe with the top 10 venues, sorted in descending way, for each province.  

```python
def return_most_common_venue(row,num_top):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top]
```

```python
num_top = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Provincia']
for ind in np.arange(num_top):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
prov_venues_sorted = pd.DataFrame(columns=columns)
prov_venues_sorted['Provincia'] = Italy_grouped['Provincia']

for ind in np.arange(Italy_grouped.shape[0]):
    prov_venues_sorted.iloc[ind, 1:] = return_most_common_venue(Italy_grouped.iloc[ind, :], num_top)

prov_venues_sorted.head()
```

```python
prov_venues_sorted.shape
```

# Cluster Province

Run k-means with 3 cluster

```python
from sklearn.cluster import KMeans

```

```python
# set number of clusters
kclusters = 5

Italy_grouped_clustering = Italy_grouped.drop('Provincia', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=1).fit(Italy_grouped_clustering)
```

Create a new DF

```python
# add clustering labels
prov_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)
prov_venues_sorted.head(2)
```

```python
Italy_merged=local.join(prov_venues_sorted.set_index('Provincia'),on='Provincia')
```

```python
Italy_merged
```

Now we have to clean the dataset from those Province that doesn't have any venue

```python
Italy_merged=Italy_merged.dropna()
Italy_merged['Cluster Labels']=Italy_merged['Cluster Labels'].astype(int)
```

```python
Italy_merged
```

Show all on map

```python
import matplotlib.cm as cm
import matplotlib.colors as colors
```

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=5)
```

```python
# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

```

```python
# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(Italy_merged['lat'], Italy_merged['long'], Italy_merged['Provincia'], Italy_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
```
